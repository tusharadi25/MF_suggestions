```
# from google.colab import drive
# drive.mount('/content/gdrive')
# df = pd.read_csv('/content/gdrive/My Drive/MF/AxisBluechipDirect.csv')
import pandas as pd
raw_git_CSV_path='https://raw.githubusercontent.com/tusharadi25/MF_suggestions/master/AxisBluechipDirect.csv'
df=pd.read_csv(raw_git_CSV_path)
```

```
print(df)
```

```
df=df.loc[df['plan_code'] == 'DG']
print(df)
```

```
data=dict(zip(df.NAV_Date,df.NAV_Value))
print(list(data.items())[:5])

  
```

```
filterDict={}
for date in data.keys():
  filterDict[date[3:]]={}
for date in data.keys():
  filterDict[date[3:]][date[:2]]=data.get(date)
print(list(filterDict.items())[:5])
```

```
import operator
for month in filterDict.keys():
  base_Nav=0
  for day in filterDict[month].keys():
    if base_Nav is 0:
      base_Nav=filterDict[month][day]
    filterDict[month][day]-=base_Nav

print(list(filterDict.items())[:5])
```

```
dist={}
for month in filterDict.keys():
  high=0
  highDate=1
  for day in filterDict[month].keys():
    if high < filterDict[month][day]:
      high = filterDict[month][day]
      highDate=int(day)
  dist[month]=highDate
print(dist)
```
